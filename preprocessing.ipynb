```python
path = 'datasets/tweets_labeled_lexicon.json'
```

```python
import pandas as pd
import re
import emoji


def clean_text(text):
    # Remove emojis
    text = emoji.replace_emoji(text)
    # Remove URLs
    # text = re.sub(r'https?://\S+', ' ', text)
    # Remove punctuation (.,?)
    # text = re.sub(r'[.,?]', ' ', text)
    # Remove quotes ('"“)
    # text = re.sub(r'[\'\"\“]', ' ', text)
    # Remove mentions
    text = re.sub(r'@[A-Za-z0-9_]+', ' ', text)
    # Remove hashtags
    text = re.sub(r'#[A-Za-z0-9_]+', ' ', text)
    # Remove newlines
    text = re.sub(r'\n', ' ', text)
    text = re.sub(r'\t', ' ', text)
    # Replace multiple spaces with a single space
    text = ' '.join(text.split())
    # Convert text to lowercase
    text = text.lower()
    return text
```

```python
df = pd.read_json(path)
df
```

```python
def has_url(text):
    match = re.search(r'https?://\S+', text)
    if match:
        return True
    return False


df['has_url'] = df['raw_content'].apply(has_url)
df.drop(df[df['has_url'] == True].index, inplace=True)
df
```

```python
df['raw_content'] = df['raw_content'].apply(clean_text)
df
```

```python
df.drop_duplicates(subset='raw_content')
```

```python
from transformers import pipeline

classifier = pipeline(
    model="lxyuan/distilbert-base-multilingual-cased-sentiments-student", 
    return_all_scores=True
)

def get_sentiment(text):
    result = classifier(text)
    return result[0]['label']

df['sentiment_automation'] = df['raw_content'].apply(get_sentiment)
```

```python
df.to_json('datasets/tweets_labeled_lexicon_cleaned.json')
```
